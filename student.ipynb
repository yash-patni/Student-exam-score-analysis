```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed column

```python
# Check if 'Unnamed: 0' exists in the columns of the DataFrame before dropping
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)
else:
    print("'Unnamed: 0' column not found in the DataFrame")

# Display the first few rows of the DataFrame to confirm
print(df.head())
```

# change weekly study hours column

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct", "5-10")
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above we have analyzed that:
the number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have s good impact on their course
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": "mean", "WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot=True)
plt.title("Relationship between Parent's Marital status and Student's score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible imapct on the student's score due to their parent's marital status
```

# Distribution of Ethnic Groups

```python
import matplotlib.pyplot as plt

# Count the number of students in each ethnic group
groupA = df.loc[df['EthnicGroup'] == "group A"].shape[0]
groupB = df.loc[df['EthnicGroup'] == "group B"].shape[0]
groupC = df.loc[df['EthnicGroup'] == "group C"].shape[0]
groupD = df.loc[df['EthnicGroup'] == "group D"].shape[0]
groupE = df.loc[df['EthnicGroup'] == "group E"].shape[0]

# List of group labels
labels = ["group A", "group B", "group C", "group D", "group E"]

# List of counts corresponding to each group
mlist = [groupA, groupB, groupC, groupD, groupE]
print(mlist)
# Plotting the pie chart
plt.pie(mlist, labels=labels, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
